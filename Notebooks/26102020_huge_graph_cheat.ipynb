# Analysis of huge Barabasi-Albert Plot using random choice of nodes
Goal: Investigate if it is reliable to investigate big graphs by only looking at a fraction of the nodes

    - Try it on a big graph and see if it is viable
    - Try it on a small graph and compare to using all the values
        - Can the result be scaled up?

```python
import igraph
import numpy as np
import random
```

```python
N = 100_000
m = 2
big_boi = igraph.Graph.Barabasi(N, m)
```

```python
# n is the number of randomly chosen nodes
n = 1_000
random_nodes = random.sample(range(0,N), n)

d_array = np.array(big_boi.degree(random_nodes))
c_array = np.array(big_boi.closeness(random_nodes))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(np.log(d_array), 1 / c_array)
```

Seems like it works decently well if you just take enough points

z value should be a good determinant of how good the approximation to the original graph is
