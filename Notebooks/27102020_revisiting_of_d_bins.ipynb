# Revisiting binning of degree values

```python
import igraph
import numpy as np
import matplotlib.pyplot as plt
import graph_functions
```

```python
g = igraph.Graph.Barabasi(40_000, 3)

d_raw = g.degree()
c_raw = g.closeness()
```

```python
# group in degree
data_dict = dict()
for d, c in zip(d_raw, c_raw):
    if d not in data_dict:
        data_dict[d] = []  
    data_dict[d].append(c)
```

```python
x,y,sdx,sdy = graph_functions.closeness_and_degree_to_x_y_bins(c_raw,d_raw)
```

```python
plt.errorbar(x,y,xerr=sdx,yerr=sdy, linestyle='None',capsize=4)
```
